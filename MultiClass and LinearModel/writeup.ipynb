## **WU4** ##
 Come up with a non-convex univariate optimization problem. Plot the function you're trying to minimize and show two runs of gd, one where it gets caught in a local minimum and one where it manages to make it to a global minimum. (Use different starting points to accomplish this.)

f = x^4-x^3-3x^2 
is a univariate non-covex function with  global minimum at (1.656, -5.248) and a local minimum (-0.906,-1.045)

```python
#graph of f.
x = np.arange(-2,3,0.001)
f = x**4 - x**3 - 3*x**2
plt.plot(x,f)
plt.title("X^4-x^3-3x^2")
```

Finding gradient: d/dx (fx) = 4x^3 - 3x^2 - 6x
Using 100 iteration and step size of 0.1

```python
#gradient descent
x, trajectory = gd.gd(lambda x: x**4 - x**3 - 3*x**2, lambda x: 4*x**3-3*x**2-6*x, 1, 100, 0.1)
plt.plot(trajectory, label = "start at 1")
x2, trajectory2 = gd.gd(lambda x: x**4 - x**3 - 3*x**2, lambda x: 4*x**3-3*x**2-6*x, -1.5, 100, 0.1)
plt.plot(trajectory2, label = "start at -1.5")
plt.legend()
plt.title("Gradient descent on x^4-x^3-3x^2")
plt.show()
```

```python
#printing out x as result of gd.
print(x)
print(x2)
```

We see that when starting at 1, gd was able to find global minimum 
But when we started at -1.5, gd got stuck in local minimum

Hint: look at WineDataBinary.words to get the id-to-word mapping. List the top 5 positive and top 5 negative and explain.

```python
print("square loss")
sq = linear.LinearClassifier({'lossFunction': linear.SquaredLoss(), 'lambda': 1, 'numIter': 100, 'stepSize': 0.5})
runClassifier.trainTestSet(sq,WineDataBinary)
print("hinge loss")
hinge = linear.LinearClassifier({'lossFunction': linear.HingeLoss(), 'lambda': 1, 'numIter': 100, 'stepSize': 0.5})
runClassifier.trainTestSet(hinge,WineDataBinary)
print("logistic loss")
log = linear.LinearClassifier({'lossFunction': linear.LogisticLoss(), 'lambda': 1, 'numIter': 100, 'stepSize': 0.5})
runClassifier.trainTestSet(log,WineDataBinary)

#retreiving words
words = WineDataBinary.words


#mapping weight to word as list of tuple
result = list(zip(log.getRepresentation(),words))
#sort by weight
result = sorted(result)

print("five most negative words")
neg = result[:5]
print(neg)


print("five most positive result")
pos = result[-5:]
print(pos)


```
